На текущий момент бот стабильно проходит линию на 14 скорости. Увеличение скорости приводят к срывам с линии. Срывы можно решать отдельным алгоритмом,но хочется понять чем обсуловлены срывы и попробовать еще поднять сокрость прохождения.

Что в итоге сделано:

* Исправлена ошибка в выборе управляющего воздействия от левого/правого PID.
* Замер min/max на основе статистических данных в димамике. Изначально клибровка выполнялась в статике и цифры были не правильные. Для сбора данных используется вращение на черной линиии. Оказалось, что сигнал на черном имеет знаяительный разброс.
* Нормализация значений датчиков по формуле (curr-min)(max-min)
* Высокие значения KD дают возможность включать двигатель на максимальную мощность при любой ошибке и плавно снижать при уменьшении ошибки. Такая реализация устранят вялую реакцию на кривых - пока двигатели разгоняются ошибки уже достагает критических значений. 


Для уменьшения времения прохождения есть два направления:

1. Менять скорость в зависимости от кривизны. На прямых увеличивать скорость. На кривых уменьшать скорость. При вылетах выполнять возврат. Нужно делать только после выявления максимума на кривых.
    * Задействовать дополнительные датчики для определния вылета
    * Сделать сокрость динамической
    * Реализовать возврат после срыва
1. Увеличивать скорость прохождения на любой кривизне.
    * Понять почему происходит срыв. Точная причина пока не ясна. Возможно мы достигли предела, который может быть связан с инерционностью двигателей. Есть варианты улучшений алгоритма, но нет уверенности, что они приведут к занчимому эффекту.
    * Реализовать автоматическое определение min/max для датчиков. Понять как бороться с разбросом в значениях датчиков для определения min/max.
    * Перевести dt на микросекунды. Сечас идет округление до миллисекунд и dt всегда равен 1 ms.
    * Сделать сглаживание входных значений для уменьшения влияния ложных значений датчиков. В целом кажется, что у нас сглаживание автоматически происходит из-за инерционности двигателей.
    * Использовать камеру для прогнозирования кривизны линии. Сложный вариант надо будет подключать компьютерное зрение.
